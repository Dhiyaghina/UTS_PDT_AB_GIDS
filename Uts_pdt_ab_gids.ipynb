```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weather_dataset.csv')
df
```

```python
column_labels = df.columns.tolist()
for stat_type in ['Jumlah data', 'Rata-rata data', 'Median', 'Mean data']:
    print(f"** {stat_type} **")
    for label in column_labels:
        column_data = df[label]
        jumlah = len(column_data)
        rata_rata = np.mean(column_data)
        median = np.median(column_data)
        mean = np.mean(column_data)

        print(f"{label} : ", end="")
        if stat_type == 'Jumlah data':
            print(jumlah)
        elif stat_type == 'Rata-rata data':
            print("{:.2f}".format(rata_rata))
        elif stat_type == 'Median':
            print("{:.2f}".format(median))
        elif stat_type == 'Mean data':
            print("{:.2f}".format(mean))
        print()
    print()
```
